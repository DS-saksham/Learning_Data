```python
import numpy as np
```

```python
X= np.random.rand(200,1)
```

```python
X.shape
```

```python
Y = 0.6*X**2 + 3*X +np.random.rand(200,1)
```

```python
import matplotlib.pyplot  as plt
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.33, random_state=42)
```

```python
plt.scatter(X,Y)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree_reg = DecisionTreeRegressor(max_depth=3,min_samples_leaf=10)
# min_sample_leaf 10 means that the leaf node cannot have less then 10 smaples
tree_reg.fit(X_train,y_train)
```

```python
predicted_value  =tree_reg.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error ,r2_score
```

```python
mae = mean_absolute_error(y_test,predicted_value)
r2 = r2_score(y_test,predicted_value)
```

```python
mae
```

```python
r2
```

```python
plt.scatter(X_test,predicted_value)
```

```python
Y_pred_for_training = tree_reg.predict(X_train)
```

```python
r2_score(y_train,Y_pred_for_training)
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(tree_reg,
              feature_names=X_train,
              class_names = y_train,
               filled =True,
               rounded = True
               )
plt.show()
```
