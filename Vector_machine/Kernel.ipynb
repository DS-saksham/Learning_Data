# Radical Basis Function(RBF) kernel

```python
# A linear SVM tries to draw a straight line or hyperplane to separte classes.
# RBF kernel maps the original data to a higher dimesnsional space where the classes become 
# linearly separable 
# THE RBF kernel does this implicitly using a mathematical trick called the kernel trick
# it doesnot compute the new dimensions directly but still behave as if it did
```
