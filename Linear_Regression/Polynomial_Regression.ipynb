```python
# You can use a linear model to fit the nonlinear data . A simple way to do this is to add 
# powers to each feature as new feature, then train the linear model on this extended set of feature
# This technique is know as polynomial regression
```

```python
# Generating the non linear data
import numpy as np
m =100
X = 6 * np.random.rand(m) -3 
y = 0.5 * X**2 + X + np.random.randn(m)
```

```python
import pandas as pd
df = pd.DataFrame({"X":X, "y":y})
```

```python
# Visualization of data
import matplotlib.pyplot as plt
import seaborn as sns
plt.scatter(X,y, color="blue", label = "Original Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

```python
# PolynomialFeatures class transform the training data by adding the square (second degree
# polynomial) of each feature in the training set as a new feature
from sklearn.preprocessing import PolynomialFeatures
poly_feature = PolynomialFeatures(degree =2 ,include_bias = True)
X_ply = poly_feature.fit_transform(X)
X[0]
X_ply[0]
```

```python
# X_ploy now contains the original feature of X plus the square of the feature . Now you can
# easily fit a linearRegression model to the extended training data
# Making a new dataframe for the modified features 
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_ply,y)
lin_reg.coef_
```
