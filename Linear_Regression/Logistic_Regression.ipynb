# Logistic Regression

```python
# Logistic regression is used to estimate the probability that an instance belongs to a particular
# class (What is the probability that a given email is a spam ?) 
# Logistic regression uses the weighted sum of the inputs to compute a logit.
#The logit is passed through the sigmoid function to get the predicted probability.
#This probability is interpreted as the likelihood of the positive class (class 1).

```

```python
# Building a classifier to detect the Iris virginica type based only on the petal width feature
import pandas as pd
import numpy as np
from sklearn import datasets
iris = datasets.load_iris() # Loading the iris dataset
list(iris.keys()) # This will give us all the name of columns
X = iris["data"][:,3:] # petal with 
# ,3: takes the colums 4 and onwards ad indexing of columns start form 0
y = (iris["target"] == 2).astype(int)
# this will change True and False into 0 and 1
```

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,y)
```
